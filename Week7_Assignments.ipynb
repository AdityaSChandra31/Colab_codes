```python
def spiral(row,column,arr):
  rowstart=0;colstart=0

  while(rowstart<row and colstart<column):

    for i in range(rowstart,row):
      print(arr[i][colstart], end=' ')

    colstart+=1

    for i in range(colstart,column):
      print(arr[row-1][i], end=' ')

    row-=1

    if(rowstart<row):
      for i in range(row-1,rowstart-1,-1):
        print(arr[i][column-1],end=' ')

      column-=1

    if(colstart<column):
      for i in range(column-1,colstart-1,-1):
        print(arr[rowstart][i],end=' ')

      rowstart+=1

matrix=[[1,2,3],
        [5,6,7],
        [9,10,11]]

row=3
column=3

spiral(row,column,matrix)
```

```python
def is_magic_square(matrix, n):
    #your code goes here...
    magic_sum = n * (n**2 + 1) // 2

    for row in matrix:
        if sum(row) != magic_sum:
            return "Not a Magic Matrix"

    for col in range(n):
        if sum(matrix[row][col] for row in range(n)) != magic_sum:
            return "Not a Magic Matrix"

    if sum(matrix[i][i] for i in range(n)) != magic_sum:
        return "Not a Magic Matrix"

    if sum(matrix[i][n - 1 - i] for i in range(n)) != magic_sum:
        return "Not a Magic Matrix"

    return "Magic Matrix"

n = int(input())
matrix = [list(map(int, input().split())) for _ in range(n)]

print(is_magic_square(matrix, n))
```

```python
def find_median(matrix, n, m):
    sorted_elements = sorted([num for row in matrix for num in row])

    median_index = (n * m) // 2

    return sorted_elements[median_index]

n, m = map(int, input().split())
matrix = [list(map(int, input().split())) for _ in range(n)]

print(find_median(matrix, n, m),end='')
```

```python
def rotate_matrix(matrix, m, n):
    rotated = [[matrix[m - 1 - row][col] for row in range(m)] for col in range(n)]

    for row in rotated:
        print(*row)

m, n = map(int, input().split())
matrix = [list(map(int, input().split())) for _ in range(m)]

rotate_matrix(matrix, m, n)
```
